# K-Means clustering

### KMeans clustering using sklearn.cluster.KMeans (For Linear Cluster boundaries)

<font>Import the required libraries.</font>

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

<font>Generate a two-dimensional dataset containing four distinct blobs and plot the datapoints.</font>

```python
from sklearn.datasets.samples_generator import make_blobs
X, y_true = make_blobs(n_samples=300, centers=4,
                       cluster_std=0.60, random_state=0)
plt.scatter(X[:, 0], X[:, 1], s=50);
```

<font>Performing KMeans using the sklearn library.Clustering is done by KMeans using the following conceptions:<br/><ol><li>The "cluster center" is the arithmetic mean of all the points belonging to the cluster.</li>
    <li>Each point is closer to its own cluster center than to other cluster centers.</li></ol></font>

<font><b>Dividing the dataset into 4 clusters</b></font>

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)
```

<font>Plotting the datapoints after clustering. Below is the representation of the different clusters formed using different colors.</font>

```python
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5);
```

<font><b>Dividing the dataset into 3 clusters and visualizing the same.</b></font>

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5);
```

<font><b>Dividing the dataset into 5 clusters and visializing the same.</b></font>

```python
kmeans = KMeans(n_clusters=5)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5);
```
